```python
import pandas as pd
pd.set_option('display.max_columns', None)

from TradeProbabilities import TradeProbabilities
```

```python
interval_type = ['1mo','1wk','1d', '1h','15m','5m']
period = [1, 1, 1, 3, 12, 24]
start_date = ['2001-01-01','2001-01-01','2001-01-01', '2021-05-01','2023-02-10','2023-02-10']
#start_date = ['2022-03-10','2022-03-10','2022-03-10','2023-02-27','2023-02-31','2023-03-31']
end_date = ['2023-02-27']*len(interval_type)
ticker = ['es=f']*len(interval_type)
pattern_type = [None]*len(interval_type)

output = []
for i in range(0,len(interval_type)):
    print(ticker[i], start_date[i], end_date[i], interval_type[i], period[i])

    # Create a StockData object with the ticker, start and end dates, interval type, period, and pattern type
    stock_data = TradeProbabilities(ticker[i],start_date[i],  end_date[i], interval_type[i], period[i])
    
    # Get the number of up and down days
    stock_data.get_up_days()
    stock_data.get_down_days()
    stock_data.get_pattern()

    # Get the pattern analytics
    pattern_analytics = stock_data.get_pattern_analytics()

    # Get the pattern summary
    pattern_summary = stock_data.get_pattern_summary()

    # Get the pattern expectations
    data = stock_data.get_pattern_expectations()
    
    output.append(data)  
    
output = pd.DataFrame(output)
```
